```python
from google.colab import userdata
from openai import OpenAI
import gradio as gr

# 🔐 Secure API key
api_key = userdata.get("ghjk")

# Azure config
endpoint = "https://kirti22scse1012550668-resource.openai.azure.com/"
deployment_name = "gpt-4.1"

# Client setup
client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}openai/deployments/{deployment_name}",
    default_query={"api-version": "2024-02-01"},
)

# 💡 Function for healthcare Q&A
def ask_healthcare_ai(question):
    try:
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {"role": "system", "content": "You are an AI expert helping healthcare analysts understand AI applications in healthcare."},
                {"role": "user", "content": question}
            ],
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"

# 🎨 Gradio UI
interface = gr.Interface(
    fn=ask_healthcare_ai,
    inputs=gr.Textbox(lines=2, placeholder="Ask about AI in healthcare..."),
    outputs="text",
    title="🏥 AI in Healthcare Assistant",
    description="Ask questions about how AI is used in healthcare (diagnosis, prediction, automation, etc.)"
)

# Launch app
interface.launch()
```
