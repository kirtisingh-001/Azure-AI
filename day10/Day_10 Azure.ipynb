```python
# Step 1: Install required library
!pip install --upgrade openai

# Step 2: Import modules
from google.colab import userdata
from openai import OpenAI

# Step 3: Load API key (saved as "kaif")
api_key = userdata.get("kaif")

# Step 4: Azure Foundry endpoint (with project path)
endpoint = "https://kirtiaif.services.ai.azure.com/api/projects/proj-default"

# Step 5: Agent details
agent_name = "email-assistant"
version = "2"
# Step 6: Create client
client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}/openai/v1"
)

# Step 7: Take user input
user_input = input("Ask your AI Email Assistant: ")

# Step 8: Send request to agent
response = client.responses.create(
    input=user_input,
    extra_body={
        "agent_reference": {
            "name": agent_name,
            "version": version,
            "type": "agent_reference"
        }
    }
)

# Step 9: Print response
print("\n AI Generated Response:\n")
print(response.output_text)
```
